```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
project_dir = "drive/My Drive/projects/predict_touchdowns/data"
in_dir = os.listdir(f"{project_dir}")

# list out datasets
print(in_dir)
```

```python
# get list of full filepaths
datasets = []
for x in in_dir:
  datasets.append(project_dir + "/" + x)


print(datasets)
```

```python
df = pd.DataFrame()

# loop through datasets and add them to our dataframe

for x in datasets:
  df = df.append(pd.read_csv(x))

# reset index --> without resetting index, we might get duplicate indices
df.reset_index(drop=True, inplace=True)
df
```

```python
df.sample(10)
```

```python
df.columns.values
```

```python
"""
now we explore the dataset.
we dont need to focus on all the features in the dataset,
we start simple with features that correlate to qb
touchdowns.
"""

# get some quarterback info
qb_features = ['season', 'passer_id', 'passer', 'pass',
            'complete_pass', 'interception',
            'sack', 'yards_gained', 'touchdown']

# group by season and passer
groupby_features = ['season', 'passer_id', 'passer']

# aggregate our groupby by sum
qb_df = df.loc[:, qb_features].groupby(groupby_features, as_index=False).sum()

qb_df.sample(10)
```

```python
"""
produce scatter plots of how different features
such as yards_gained, complete pass, interceptions
and more correlate to touchdowns.
"""

for y in ['yards_gained', 'complete_pass',
          'pass', 'interception','sack']:
  sns.regplot(data=qb_df, x='touchdown', y=y)
  plt.title(f"touchdowns and {y}")
  plt.show()
```

```python
"""
from the scatter plots, we can tell
that knowing yards gained, completed passes, attempted
passes, interceptions, and sacks all correlate with
touchdowns. Check if this is the case with touchdowns
for next season.
"""

df_copy = qb_df.copy()

df_copy['season'] = df_copy['season']+1

# merge our copy back onto the original dataframe with a left join on
# season, passer_id, and passer

new_qb_df = qb_df.merge(df_copy, on=['season', 'passer_id', 'passer'],
                        suffixes=('', '_prev'),
                        how='left')
new_qb_df.sample(10)
```

```python
"""
Find what correlates with next season touchdowns
based on the following set of scatter plots:
"""

for y in ['touchdown_prev','yards_gained_prev',
          'complete_pass_prev', 'pass_prev',
          'interception_prev','sack_prev']:
  sns.regplot(data=new_qb_df, x='touchdown', y=y)
  plt.title(f"touchdowns and {y}")
  plt.show()
```

```python
"""
We can see that there are year to year correlations with touchdowns.
Based on this information, we can use previous season performance to estimate
touchdowns thrown for the next season.
"""
```

```python
"""
Make ML model --> use linear regression
"""

from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from scipy.stats import pearsonr
```

```python
"""
Regression example: train on 2-3 seasons, use next season to test.
"""

features= ['pass_prev', 'complete_pass_prev',
            'interception_prev', 'sack_prev',
            'yards_gained_prev', 'touchdown_prev']

# drop null values

model_data = (new_qb_df.dropna(subset=features+['touchdown']))

# train on 2015, 2016, 2017, 2018, 2019 data

# test on 2020, 2021, 2022, 2023

train_data = model_data[ model_data['season'].isin([2015, 2016, 2017, 2018, 2019, 2020]) ]

test_data = model_data[model_data['season'].isin([2021, 2022, 2023])]

model = LinearRegression()
model.fit(train_data.loc[:, features], train_data['touchdown'])

preds = model.predict(test_data.loc[:, features])

preds = pd.Series(preds, index=test_data.index)

test_data['preds'] = preds
```

```python
# Run some statistics to examine the quality of our predictions

rmse = mean_squared_error(test_data['touchdown'], test_data['preds']) ** 0.5
r2 = pearsonr(test_data['touchdown'], test_data['preds'])[0] ** 2
print(f"RMSE: {rmse}\nr2: {r2}")
```

```python
# visualize outputs
sns.regplot(data=test_data, x='touchdown', y='preds')
plt.title('touchdown and predictions')
plt.show()
```
